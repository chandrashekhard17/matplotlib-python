```python
#2...
"""A scatter plot is a type of plot or mathematical diagram that uses Cartesian coordinates to display values for typically two variables for a set of data. The data is displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis (x-axis) and the value of the other variable determining the position on the vertical axis (y-axis). Scatter plots are used to observe relationships between variables.

Here's the code to generate the data for x and y and create a scatter plot with a title and labels for the x and y axe"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X values')
plt.ylabel('Y values')

# Show plot
plt.show()
```

```python
#3...
"""The subplot() function in Matplotlib allows you to create multiple plots (subplots) within a single figure. It’s useful when you want to display several related plots side by side."""
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for four lines
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(12, 6))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')

# Add common labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
#4....
import matplotlib.pyplot as plt
import numpy as np

# Given data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='skyblue')
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.title("Company Profits")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='salmon')
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.title("Company Profits (Horizontal)")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(6, 4))
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"], notch=True, vert=False)
plt.xlabel("Values")
plt.ylabel("Boxes")
plt.title("Box Plot")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
