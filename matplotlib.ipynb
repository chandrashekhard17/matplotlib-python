```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
plt.figure(figsize=(10,5))
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,color='red',alpha=.9)
plt.xlabel("this is the x label of the 3d projection")
plt.ylabel('This is the y label of the 3d projection')
plt.show()
```
